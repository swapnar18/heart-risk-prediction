```python
import os
import glob
from tkinter import filedialog
import tkinter as tk
from PIL import ImageTk, Image
import shutil
import os

class ImageViewer(object):
    def __init__(self):

        self.root = tk.Tk()
        self.root.state('zoomed')
        self.width = self.root.winfo_screenwidth()
        self.height = self.root.winfo_screenheight()
        self.images = None
        self.current_image = None
        self.current_img_num=0

        img, wh, ht = self.open_file()
        self.canvas = tk.Canvas(self.root, width=wh, height=ht, bg='black')
        self.canvas.pack(expand=tk.YES)
        self.image_on_canvas = self.canvas.create_image(self.width/2, self.height/2, anchor=tk.CENTER, image=img)

        b = tk.Button(self.root, text='next', command=self.next_image)
        b.place(x=50, y=50)
#         b = tk.Button(self.root, text='back', command=self.previous_image)
#         b.place(x=100, y=100)
        m = tk.Button(self.root, text='move', command=self.move_image)
        m.place(x=50, y=80)
        
        p = tk.Button(self.root, text='back', command=self.prev_image)
        p.place(x=50, y=110)

        self.root.mainloop()

    def open_file(self):
        openfile = filedialog.askopenfilename(initialdir='/', title='Select image', filetypes=(('jpeg files', '*.jpg'), ('all files', '*.*')))
        self.images = glob.glob(os.path.dirname(openfile) + '/*.jpg') + glob.glob(os.path.dirname(openfile) + '/*.png') \
                        + glob.glob(os.path.dirname(openfile) + '/*.jpeg')
        self.root.title(openfile)
        img = Image.open(openfile)
        self.current_image = openfile
        self.current_img_num = 0
        img.thumbnail((self.width, self.height), Image.ANTIALIAS)
        img = ImageTk.PhotoImage(img)

        return img, self.width, self.height

    def next_image(self):
        if not self.images:
#             print('in 1')
            img, wh, ht = self.open_file()
            self.current_img_num = 0
        else:
#             print('in 2')
#             print(self.images)
            self.current_img_num +=1
            image = self.images[self.current_img_num]
#             print(self.current_img_num)
#             print(image)
            self.current_image = image
            self.root.title(image)
            img = Image.open(image)
            img.thumbnail((self.width, self.height), Image.ANTIALIAS)
            img = ImageTk.PhotoImage(img)

        self.canvas.itemconfigure(self.image_on_canvas, image=img)
        self.canvas.config(width=self.width, height=self.height)
        try:
            self.canvas.wait_visibility()
        except tk.TclError:
            pass
        
    def prev_image(self):
        if self.current_img_num == 0:
            pass
            self.current_img_num = 0
        else:
#             print('in 2')
#             print(self.images)
            self.current_img_num -=1
            image = self.images[self.current_img_num]
#             print(self.current_img_num)
#             print(image)
            self.current_image = image
            self.root.title(image)
            img = Image.open(image)
            img.thumbnail((self.width, self.height), Image.ANTIALIAS)
            img = ImageTk.PhotoImage(img)

        self.canvas.itemconfigure(self.image_on_canvas, image=img)
        self.canvas.config(width=self.width, height=self.height)
        try:
            self.canvas.wait_visibility()
        except tk.TclError:
            pass
        
    def move_image(self):
#         print(self.current_image)
        self.current_image = self.current_image.replace('\\','/')
        cwd = os.getcwd()
        dst = 'F:/Marriage Selected Pics'
#         print(self.current_image.split('/'))
#         print(self.current_image.split('/')[-1])
        if not os.path.exists(dst+'/'+self.current_image.split('/')[-1]):
            shutil.copy(self.current_image, os.path.join(dst, self.current_image.split('/')[-1]))
        else:
            img_nm = self.current_image.split('/')[-1].split('.')[0]
            img_ext = self.current_image.split('/')[-1].split('.')[1]
            if not os.path.exists(dst+'/'+img_nm+'_1.'+img_ext):
                shutil.copy(self.current_image, os.path.join(dst, img_nm+'_1.'+img_ext))
            else:
                shutil.copy(self.current_image, os.path.join(dst, img_nm+'_2.'+img_ext))
        try:
            self.canvas.wait_visibility()
        except tk.TclError:
            pass
ImageViewer()
```

```python
import os
import glob
from tkinter import filedialog
import tkinter as tk
from PIL import ImageTk, Image
class Display(object):

    def __init__(self):
        self.images = [];
        self.imgIndex = 0;
        self.master= tk.Tk()
        self.framePhoto = Frame(self.master, bg='gray50',relief = RAISED, width=800, height=600, bd=4)
        self.framePhoto.pack();
        prevBtn = Button(self.framePhoto, text='Previous', command=lambda s=self: s.getImgOpen('prev')).place(relx=0.85, rely=0.99, anchor=SE)
        nextBtn = Button(self.framePhoto, text='Next', command=lambda s=self: s.getImgOpen('next')).place(relx=0.90, rely=0.99, anchor=SE)
        #prevBtn.pack();
        #nextBtn.pack();
        self.getImgList('test_2/test_2','.bmp')
        mainloop()

    def getImgList(self, path, ext):
        imgList = [os.path.normcase(f) for f in os.listdir(path)]
        imgList = [os.path.join(path, f) for f in imgList if os.path.splitext(f)[1] == ext]
        self.images.extend(imgList)
        #print self.images

    def getImgOpen(self,seq):
        print('Opening ',seq)
        if seq=='ZERO':
            self.imgIndex = 0
        elif (seq == 'prev'):
            if (self.imgIndex == 0):
                self.imgIndex = len(self.images)-1
            else:
                self.imgIndex -= 1
        elif(seq == 'next'):
            if(self.imgIndex == len(self.images)-1):
                self.imgIndex = 0
            else:
                self.imgIndex += 1

        self.masterImg = Image.open(self.images[self.imgIndex]) 
        self.master.title(self.images[self.imgIndex])
        self.masterImg.thumbnail((400,400))
        self.img = ImageTk.PhotoImage(self.masterImg)
        label = Label(image=self.img)
        label.image = self.img # keep a reference!
        label.pack()
        label.place(x=100,y=100)
        #self.lbl['image'] = self.img
        return    
d = Display();
d.getImgOpen('next')
```

```python
from tkinter import *

class SlideShowGUI:

    def __init__(self, parent, counter = 0):
        
        self.counter = counter
        
        self.images_list = ["smiley.png", "carrot.png", "bunny.jpg"]
        self.photo = PhotoImage(file = self.images_list[self.counter])

        self.b1 = Button(parent, text = "", image = self.photo, bg = "white")
        self.b1.grid(row = 0, column = 0)
        
        back = Button(parent, width = 2, anchor = W, text = "<", relief = RIDGE, command = self.previous_image)
        back.grid(row = 1, column = 0)
        
        forward = Button(parent, width = 2, anchor = E, text = ">", relief = RIDGE, command = self.next_image)
        forward.grid(row = 1, column = 1)

    def previous_image(self):
        self.counter -= 1
        self.photo.configure(file = self.images_list[self.counter])        

    def next_image(self):
        self.counter += 1
        self.photo.configure(file = self.images_list[self.counter])  

#main routine

if __name__ == "__main__":
    root = Tk()
    root.title("hello")
    slides = SlideShowGUI(root)
    root.mainloop()
```
